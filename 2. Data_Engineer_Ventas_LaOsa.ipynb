## Damos formato a las columnas Id y Nombre del producto
* Se separa el código de cada producto del nombre del producto en dos columnas diferentes
* Usamos la expresión regular para extraer el código sin espacios ni `[]`
* Reorganizamos el orden de las columnas, anteponiendo de primero las dos columnas creada con el `Id` y el `nombre`
* Eliminamos la columna `Id_Prod` inicial

```python
# Se cambia la columna `Id_Prod` por tipo de dato `category`
dfventas_2021["Id_Prod"]=dfventas_2021["Id_Prod"].astype("category")
```

```python
# Usamos la expresión regular para extraer el código y el nombre de cada producto
patron = r'\[(.*?)\]\s*(.*)'#expresion regular a extraer de la columna

# Se crean dos columnas nuevas donde irá el Id del producto (Id_Prod_2) y el Nombre dle producto (Nom_Prod)
dfventas_2021[['Id_Prod_2', 'Nombre_Prod']] = dfventas_2021['Id_Prod'].str.extract(patron)

#Verificamos las columnas creadas y la extraccion del codigo
dfventas_2021.info()
```

* Reorganizamos el orden de las columnas, anteponiendo de primero las dos columnas creada con el `Id` y el `nombre`

```python
# Obtén una lista con los nombres de las columnas  y el orden que queremos que tengan en el DataFrame
nuevas_columnas = ['Id_Prod_2', 'Nombre_Prod'] + list(dfventas_2021.columns[:-2])
```

```python
# Reorganizamos el DataFrame 
dfventas_2021 = dfventas_2021.reindex(columns=nuevas_columnas)
dfventas_2021.head(3)
```

* Eliminamos la columna `Id_Prod` inicial

```python
# Eliminar columna Id_Prod
dfventas_2021=dfventas_2021.drop("Id_Prod",axis=1)
```
